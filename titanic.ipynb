```python
# Analiza Fare w zbiorze Titanic

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# upload danych
df = pd.read_csv("titanic.csv")

# Podstawowe statystyki opisowe
print("Statystyki opisowe zmiennej Fare:")
print(df["Fare"].describe())

# Wykres pudełkowy 
plt.figure(figsize=(10, 2))
sns.boxplot(x=df["Fare"])
plt.title("wykres pudełkowy zmiennej Fare")
plt.xlabel("Fare")
plt.show()

# Histogram z estymacją gęstości KDE
plt.figure(figsize=(8, 4))
sns.histplot(df["Fare"], bins=50, kde=True)
plt.title("Histogram zmiennej Fare")
plt.xlabel("Fare")
plt.ylabel("Liczba pasażerów")
plt.show()

# sekcja komentarzy
print("\nWnioski:")
print("- Średnia opłata za bilet wynosi około 32.20.")
print("- Mediana (50%) to około 14.45, co sugeruje skośność prawostronną.")
print("- Występują wartości odstające – najwyższa opłata to ponad 500.")
print("- Większość pasażerów zapłaciła mniej niż 30.")
```
